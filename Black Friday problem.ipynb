```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
x = ds[['Gender','Age','Occupation','City_Category','Stay_In_Current_City_Years','Marital_Status','Product_Category_1','Product_Category_2','Product_Category_3']]
y= ds[['Purchase']]
```

# Encoding

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
```

```python
x= x.apply(le.fit_transform)
```

```python
x.Gender= pd.to_numeric(x.Gender)
x.Age= pd.to_numeric(x.Age)
x.Occupation= pd.to_numeric(x.Occupation)
x.City_Category= pd.to_numeric(x.City_Category)
x.Stay_In_Current_City_Years= pd.to_numeric(x.Stay_In_Current_City_Years)
x.Marital_Status= pd.to_numeric(x.Marital_Status)
x.Product_Category_1= pd.to_numeric(x.Product_Category_1)
x.Product_Category_2= pd.to_numeric(x.Product_Category_2)
x.Product_Category_3= pd.to_numeric(x.Product_Category_3)
```

# Splitting data

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.3)
```

```python
print(x_train.shape)
print(y.shape)
```

```python
reg= LinearRegression()
reg.fit(x_train,y_train)
```

```python
print(reg.intercept_)
print(reg.coef_)
```

```python
y_predict = reg.predict(x_test)
```

```python
y_predict
```

```python
df= pd.DataFrame({'Actual':[y_test],'Predicted':[y_predict]})
df
```

```python
print('MAE', metrics.mean_absolute_error(y_test,y_predict))
print('MSE', metrics.mean_squared_error(y_test,y_predict))
print('RMSE', np.sqrt(metrics.mean_squared_error(y_test,y_predict)))
model_r2= reg.score(x_train,y_train)
print('r2 score:',model_r2)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regr1= DecisionTreeRegressor()
dt_regr1= regr1.fit(x_train,y_train)
y_predict1= regr1.predict(x_test)
print('RMSE',np.sqrt(metrics.mean_squared_error(y_test,y_predict1)))
```

# Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingRegressor
regr2= GradientBoostingRegressor()
gbregr= regr2.fit(x_train,y_train)
y_predict2= regr2.predict(x_test)
print('RMSE',np.sqrt(metrics.mean_squared_error(y_test,y_predict2)))
```

```python
print("Accuracy Score of Gradient Boosting on testset",gbregr.score(x_test,y_test)*100)
```
